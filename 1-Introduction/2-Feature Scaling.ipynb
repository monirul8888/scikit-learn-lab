## Normalization
### Min Max Scaling


## 📌 Min-Max Scaling Formula

Min-Max scaling is used to normalize data into a fixed range (usually 0 to 1).

$$
X' = a + \frac{(X - X_{\min})(b - a)}{X_{\max} - X_{\min}}
$$

---

## 📖 Explanation

- \(X\) → original value  
- \(X_{\min}\) → minimum value in the dataset  
- \(X_{\max}\) → maximum value in the dataset  
- \(a\) → lower bound of the new scale (commonly 0)  
- \(b\) → upper bound of the new scale (commonly 1)  
- \(X'\) → normalized (scaled) value  

---

## 🔥 Special Case (0 to 1 Scaling)

When \(a = 0\) and \(b = 1\), the formula becomes:

$$
X' = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
$$

---

## 🧠 Intuition

This formula **rescales all values proportionally** so that:
- Minimum value becomes 0
- Maximum value becomes 1
- All other values stay in between
