## 📌 StandardScaler (Z-score Normalization)

StandardScaler transforms data so that it has **mean = 0** and **standard deviation = 1**.

---

## 📐 Formula

$$
X' = \frac{X - \mu}{\sigma}
$$

---

## 📖 Explanation

- \(X\) → original value  
- \(\mu\) → mean of the feature  
- \(\sigma\) → standard deviation of the feature  
- \(X'\) → standardized value  

---

## 🧠 Intuition

StandardScaler converts values into **how many standard deviations away from the mean** they are.

---

## 🔥 Example

If:
- \(X = 80\)  
- \(\mu = 70\)  
- \(\sigma = 10\)

Then:

$$
X' = \frac{80 - 70}{10} = 1
$$

👉 Meaning: 80 is **1 standard deviation above the mean**

---

## ⚖️ Key Idea

- Center data around 0  
- Scale variance to 1  
- Keeps distribution shape but standardizes scale
